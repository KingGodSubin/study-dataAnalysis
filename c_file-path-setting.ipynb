## 여러 파일을 한번에 로드하는 방법

```python
# csv로 끝나는 모든 파일 가져오기
file_csv = glob("data/store/*.csv")
file_csv
```

```python
file_list = []
for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name, low_memory=False)
    file_list.append(temp)
```

```python
len(file_list)
```

```python
df = pd.concat(file_list)
df.shape
```
