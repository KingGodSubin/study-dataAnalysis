# 분석과 시각화를 위한 tidy data 만들기
* https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf

```python
df_raw.columns
```

```python
df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"], var_name="기간", value_name="(단위:백만원)")
df.shape
```

```python
df.sample(5)
```

```python
df.info()
```

# 데이터 전처리

## 기간에서 연도를 분리하기

```python
int("2019.4/4.p".split('.')[0])
```

```python
df["연도"] = df["기간"].map(lambda x : int(x.split('.')[0]))
df.head()
```

```python
"2019.4/4.p".split('.')[1].split('/')[0]
```

```python
# 기간에서 분기만 불리하기
df["분기"] = df["기간"].map(lambda x : int(x.split('.')[1].split('/')[0]))
df.head()
```

## 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# - 문자를 결측치로 변경하고 float 타입으로 변경하기

df["(단위:백만원)"] = df["(단위:백만원)"].replace("-", np.nan).astype(float)
df.head()
```

## 필요없는 데이터 제거하기

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거합니다.
df = df[(df["국가(대륙)별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
```

```python
# 결측치 보기
df.isnull().sum()
```

# K-Beauty 시각화

## 전체 상품군 판매액

```python
# 판매유형별 데이터는 일부 기간에는 "계"만 존재하기 때문에
# 판매유형별 == "계" 데이터만 가져와서 봐야지
# 평균 값을 구하는 그래프에서 올바른 값을 표현할 수 있습니다.
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
# 그래프를 보면 2014년도부터 2018년도까지 계속 판매액이 증가하고 있음을 볼 수 있다.
sns.lineplot(data=df_total, x="연도", y="(단위:백만원)")
```

* legend를 그래프 밖에 그리기: <a href="https://www.delftstack.com/ko/howto/seaborn/seaborn-legend-position/">matplotlib - Move legend outside figure in seaborn tsplot - Stack Overflow</a>

```python
# 연도, 판매액 lineplot으로 그리고 상품군별로 다른 색상으로 표시하기

sns.lineplot(data=df_total, x="연도", y="(단위:백만원)", hue="상품군별")

# 범주가 그래프 안에 곂쳐서 보여지기 때문에 밖으로 빼기 위한 코드
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위에 그린 그래프를 자세히 보기 위해 서브플롯으로 표시하기
sns.relplot(data=df_total, x="연도", y="(단위:백만원)", hue="상품군별", kind="line", col="상품군별", col_wrap=2)
```

```python
# isin을 사용해 화장품만 제외를 하고 df_sub 이라는 변수에 담습니다.

df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
# 연도별 판매액을 상품군별로 replot을 활용해 서브플롯으로 그려보기

sns.relplot(data=df_sub, x="연도", y="(단위:백만원)", hue="상품군별", col="상품군별", col_wrap=4, kind="line")
```

## 화장품의 온라인쇼핑 해외직접판매액

```python
# df_cosmetic 이라는 변수에 상품군별이 화장품인 데이터만 가져오기

df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique()
```

```python
# 연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상으로 표현해 보기

plt.figure(figsize=(15, 4))
sns.lineplot(data=df_cosmetic, x="연도", y="(단위:백만원)", hue="분기")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화 하기

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="(단위:백만원)")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하고 "국가(대륙)별" 다른 색상으로 표시하기
# 중국이 가장 판매액이 증가한 것을 볼 수 있음

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="(단위:백만원)", hue="국가(대륙)별")
```

```python
# 중국으로 인해 다른 나라들의 증가량이 잘 보이지 않기 때문에 중국만 제외시키고 다시 그래프를 그림

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], x="기간", y="(단위:백만원)", hue="국가(대륙)별")
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하고 "판매유형별"로 다른 색상으로 표시하기

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()
sns.lineplot(data=df_sub, x="기간", y="(단위:백만원)", hue="판매유형별")
```

## 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
# df_fashion 이라는 변수에 의류 데이터만 가져와 따로 담아주기
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
# "의료 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화하고
# "국가(대륙)별"로 다른 색상으로 표시하기

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="(단위:백만원)", hue="국가(대륙)별")
```

```python
# "의료 및 패션관련 상품" 판매액에 대한 기간별 금액 데이터 시각화하고
# "판매유형별"로 다른 색상으로 표시하기

df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] != "계")].copy()

plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="(단위:백만원)", hue="판매유형별")
```

## 데이터 집계하기

```python
df_fashion["판매유형별"].value_counts()
```

```python
# 피봇테이블로 "국가(대륙)별", "연도"별 합계 금액을 표 형태로 구하기

result = df_fashion.pivot_table(index="국가(대륙)별", columns="연도", values="(단위:백만원)", aggfunc="mean")
result
```

## 연산결과를 시각적으로 보기

```python
# 피봇테이블로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기
sns.heatmap(result, cmap="Blues")
```
