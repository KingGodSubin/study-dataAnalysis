# DataFrame

```python
# () 
# pd.DataFrame?
# pd.DataFrame(shift+tab+tab)

df = pd.DataFrame(
    {"a" : [4, 5, 6, 4],
     "b" : [7, 8, 9, 9],
     "c" : [10, 11, 12, 10]},
    index = [1, 2, 3, 4]
)
df
```

# Reshaping
# sort_values, drop

```python
df.sort_values("a", ascending=False)
```

```python
df = df.drop(["c"], axis=1)
df
```

# Group Data
# Groupby, pivot_table

```python
# mean -> a  b  4   7 + 9 = 16, 16 / 2 = 8  8 
df.groupby(["a"])["b"].agg(["mean", "sum", "count"])
```

```python
df.groupby(["a"])["b"].describe()
```

```python
df
```

```python
pd.pivot_table(df, index="a", values="b", aggfunc="sum")
```
